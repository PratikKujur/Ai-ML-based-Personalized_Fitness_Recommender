```python
cat_features=[x for x in df_train.columns if df_train[x].dtype=='object']
cat_features.remove('Name of Exercise')
cat_features.remove('Benefit')
print(cat_features)
```

```python
num_features = [x for x in df_train.columns if x not in cat_features + ['Name of Exercise', 'Benefit']]
num_features
```

```python
target_col='Name of Exercise'
```

```python
df_cat=df_train[cat_features+[target_col]]
```

```python
df_col=df_train[cat_features+[target_col]]
df_col.head()
```

```python
# perform encoding on categorical features and perform feature selection
for col in cat_features:
    print(f"Column: {col}")
    print(df_col[col].value_counts())
    print("\n")
```

```python
df_col.columns
```

```python
for col in cat_features:
    mapping_dict={}
    for x,y in zip(df_col[col].unique(), range(len(df_col[col].unique()))):
        mapping_dict[x]=y
    df_col[col]=df_col[col].map(mapping_dict)
df_col.head()
```

```python
mapping_dict={}
for x,y in zip(df_col['Name of Exercise'].unique(), range(len(df_col['Name of Exercise'].unique()))):
    mapping_dict[x]=y
df_col['Name of Exercise']=df_col['Name of Exercise'].map(mapping_dict)
df_col.head()
```

```python
#perform feature selection using chi2 test with categorical features
from sklearn.feature_selection import chi2  
X = df_col.drop(columns=[target_col])
y = df_col[target_col]
chi2_scores, p_values = chi2(X, y)
chi2_scores
```

```python
df_cat['Gender'].value_counts().keys().tolist()
```

```python
chi2_df = pd.DataFrame({'Feature': X.columns, 'Chi2_Score': chi2_scores, 'P_Value': p_values})
chi2_df.sort_values(by='Chi2_Score', ascending=False)
chi2_df['column_name']=None
for i in range(len(chi2_df)):
    chi2_df['column_name'][i]=df_cat[chi2_df['Feature'][i]].value_counts().keys().tolist()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(chi2_df)

```

```python
final_cat_features=['Gender','diet_type','cooking_method','Equipment Needed','Difficulty Level','Body Part']
```

```python
final_var_features=['Age','Water_Intake (liters)','Height (m)','Weight (kg)','Session_Duration (hours)','Workout_Frequency (days/week)','Experience_Level','Daily meals frequency']
final_features=final_cat_features+final_var_features
final_features.sort()
final_features
```

```python
for x in final_features:
    if x not in df_train.columns:
        print(x)
```

```python
df_train_final=df_train[final_features+[target_col]]
df_train_final.head()
```

```python
label_mapping = {v: i for i, v in enumerate(df_train_final[target_col].dropna().unique())}
X_final = df_train_final.drop(columns=[target_col]).copy()
y_mapped = df_train_final[target_col].map(label_mapping)

mask = y_mapped.notna()
X_final = X_final.loc[mask].reset_index(drop=True)
y_final = y_mapped.loc[mask].astype(int).reset_index(drop=True)
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)
```

```python
model_params = {k: v for k, v in model_params.items() if v is not None}
```

```python
def preprocess_for_xgb(df):
    df = df.copy()
    for col in final_cat_features:
        mapping_dict = {x: i for i, x in enumerate(df_train[col].unique())}
        df[col] = df[col].map(mapping_dict)
    return df

def preprocess_target(df):
    df = df.copy()
    mapping_dict = {x: i for i, x in enumerate(df_train['Name of Exercise'].unique())}
    df['Name of Exercise'] = df['Name of Exercise'].map(mapping_dict)
    return df
```

```python
import xgboost as xgb
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer

xgb_model = xgb.XGBClassifier(**model_params)
preprocess_transformer = FunctionTransformer(preprocess_for_xgb, validate=False)
pipeline = Pipeline(steps=[
    ('dropna', FunctionTransformer(lambda x: x.dropna(), validate=False)),
    ('preprocess', preprocess_transformer),
    ('model', xgb_model)
])
pipeline.fit(X_train, y_train)
```
